```python
PATH = 'C:\Program Files (x86)\chromedriver.exe' #path name is dependent to each computer
driver = webdriver.Chrome(PATH)
df = pd.DataFrame(columns=['Name','Quotes']) # creates a master dataframe to store our data later

lists = ["Abigail", "Akira", "Akuma", "Alex", "Balrog", "Birdie", "Blanka", "Cammy", "Chun-Li", "Cody", "Dan", "Dhalsim", "E._Honda", "Ed", "F.A.N.G", "Falke", "G", "Gill#SFV", "Guile", "Ibuki", "Juri_Han", "Kage", "Karin", "Ken_Masters#SFV", "Kolin", "Laura", "Lucia", "Luke", "M._Bison#SFV", "Menat", "Charlie_Nash", "Necalli", "Oro", "Poison", "R._Mika", "Rashid", "Rose#SFV_", "Ryu#SFIII", "Sagat", "Sakura", "Seth", "Urien", "Vega", "Zangief", "Zeku"]
characters = []
quotes = []

#This code is a framework which I think can be used for other websites as long as the structure is consistent
#In this example, although the code can access the list of the websites specified, each webpage has a different strucuture which creates some error in our results
#Nonetheless, I feel this can be a base for my future projects
for a in lists:
    url = 'https://streetfighter.fandom.com/wiki/'+a+'' 
    #iterating through the lists that we have specified as the mainpage is the same
    driver.get(url)
    names = driver.find_elements(By.XPATH,'//*[@id="firstHeading"]')
    facts = driver.find_elements(By.XPATH,'/html/body/div[4]/div[3]/div[4]/main/div[3]/div/div/dl[1]/dd[1]/span')
    #for this Xpath, this might be different to each webpage.
    for p in range(len(names)):
        characters.append(names[p].text)
    
    for s in range(len(facts)):
        quotes.append(facts[s].text)
        
data_tuples = list(zip(characters,quotes))   
new_df = pd.DataFrame(data_tuples, columns=['Name','Quotes'])
df = df.append(new_df) # appends to master dataframe
    
driver.close()

df.to_csv('character_quotes.csv', encoding='utf-8', index=False, header = True)
#utf-8 encoding can ensure that foreign characters(such as Japanese Kanji)can be identified properly
```
